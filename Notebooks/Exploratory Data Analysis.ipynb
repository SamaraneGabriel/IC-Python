```python
df_numeric = df.select_dtypes(include=[np.number])
columns = df_numeric.columns

 
plt.figure(figsize = (15,15))

plt.suptitle("Análise KDE das colunas")


#Fazendo a análise de densidade das primeiras 20 colunas numéricas

for i in range(len(columns) - 25):
    plt.subplot(6,5,i + 1)
    sns.kdeplot(x=df[columns[i]], shade = True)
    plt.xlabel(columns[i])
    plt.tight_layout()
```

```python
#PRIMEIRA OBSERVAÇÃO DE DADO POSSIVELMENTE ERRADO

#CANDIDATO COM -1 DIAS NA EMPRESA
df[df['dias_permanencia'] < 0]
```

```python
#Filtrando o outlier 
df_filtrado = df[df['dias_permanencia'] >= 0]
```

```python
#NENHUM VALOR DUPLICADO NO DATA FRAME
df.duplicated().sum()
```

```python
#A partir de agora vamos utilizar o df_filtrado, que nao possui o outlier 

#Checando NULIDADE das tuplas na tabela

nulos = df_filtrado.isna().sum()

#Função que retorna dataframe sem colunas que possuam 30% ou mais de elementos nulos 
def remove_nulos(nulls, df):
    removed_cols = []
    for i in nulls.keys():
        if nulls[i] >= df.shape[0] * 0.3:
            removed_cols.append(i)
    return df.drop(removed_cols, axis = 1)

df_filtrado = remove_nulos(nulos, df_filtrado)
```

```python
#Analisando a relação linear entre as variáveis (numéricas)

corr = df_numeric.corr()
corr.head(5)
```

```python
#Gerando um mapa de calor para corr

sns.heatmap(data=corr, cmap='CMRmap')
```

```python
#Calculando variáveis com maior quantidade de valores únicos, ou seja, variáveis contínuas 
columns_num = df_filtrado.select_dtypes(include=[np.number]).columns

columns_filtradas = []

#Filtra apenas variaveis que possuem pelo menos 950 valores distintos, favorecendo gráficos de dispersão
for col in columns_num: 
    if df[col].nunique() >= 950:
        columns_filtradas.append(col)

columns_filtradas
```

```python
#Modelo de gráficos relacionados a dominância de um funcionário com base em outras variáveis numéricas
plt.figure(figsize = (15,15))
for i in range (len(columns_filtradas)):
    plt.subplot(4,4,i+1)
    sns.scatterplot(data = df_filtrado, x='dominancia',y=columns_filtradas[i], )
    plt.tight_layout()
```

```python
#Selecionando todas as variáveis categoricas do modelo
columns_cat = df_filtrado.select_dtypes(include=['object', 'category']).columns

#Comparando com o perfil para relacionar com dominancia

plt.figure(figsize = (15,10))

sns.barplot(data=df_filtrado,x='perfil',y='dominancia',ec='black',palette="tab10")
plt.tight_layout()
```

```python
#Explorando um pouco mais a fundo as relações entre dominancia e outras variáveis

plt.figure(figsize=(15,10))

sns.scatterplot(data=df_filtrado,x='dominancia',y='proatividade',hue='perfil',s=100)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.3)
```
